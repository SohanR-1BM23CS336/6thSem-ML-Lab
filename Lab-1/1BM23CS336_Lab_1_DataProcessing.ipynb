```python
import pandas as pd

# i. Load .csv file into the data frame
df = pd.read_csv('housing.csv')
print("CSV file loaded successfully.")
display(df.head())
```

```python
# ii. Display information of all columns
print("\nInformation of all columns:")
df.info()
```

```python
# iii. Display statistical information of all numerical columns
print("\nStatistical information of all numerical columns:")
display(df.describe())
```

```python
# iv. Display the count of unique labels for 'Ocean Proximity' column
print("\nCount of unique labels for 'Ocean Proximity' column:")
display(df['ocean_proximity'].value_counts())
```

```python
# v. Display which attributes (columns) in a dataset have missing values count greater than zero
print("\nColumns with missing values:")
missing_values = df.isnull().sum()
columns_with_missing_values = missing_values[missing_values > 0]
if not columns_with_missing_values.empty:
    display(columns_with_missing_values)
else:
    print("No columns with missing values found.")
```
