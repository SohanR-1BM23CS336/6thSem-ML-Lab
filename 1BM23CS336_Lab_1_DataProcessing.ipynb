```python
import requests

url = 'https://raw.githubusercontent.com/plotly/datasets/master/diabetes.csv'
response = requests.get(url)

if response.status_code == 200:
    with open('diabetes.csv', 'wb') as f:
        f.write(response.content)
    print("Diabetes dataset downloaded successfully as 'diabetes.csv'.")
else:
    print(f"Failed to download dataset. Status code: {response.status_code}")
```

```python
import pandas as pd

df_diabetes = pd.read_csv('diabetes.csv')
print("Diabetes dataset loaded into df_diabetes successfully.")
display(df_diabetes.head())
```

```python
print("1. First 5 rows of df_diabetes:")
display(df_diabetes.head())

print("\n2. Information of all columns in df_diabetes:")
df_diabetes.info()

print("\n3. Statistical information of numerical columns in df_diabetes:")
display(df_diabetes.describe())

print("\n4. Missing values in df_diabetes:")
display(df_diabetes.isnull().sum())
```

```python
import numpy as np

# 1. Identify columns where a value of 0 is physiologically implausible and likely represents a missing value
columns_with_potential_zeros_as_missing = ['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']

print("Original value counts of 0s in identified columns:")
for col in columns_with_potential_zeros_as_missing:
    if 0 in df_diabetes[col].unique():
        print(f"{col}: {df_diabetes[col].value_counts()[0]} zeros")

# 2. Replace the 0 values with NaN
for col in columns_with_potential_zeros_as_missing:
    df_diabetes[col] = df_diabetes[col].replace(0, np.nan)

print("\nAfter replacing 0s with NaN, checking for NaN counts:")
display(df_diabetes[columns_with_potential_zeros_as_missing].isnull().sum())

# 3. Impute the NaN values in these columns using the median
for col in columns_with_potential_zeros_as_missing:
    median_val = df_diabetes[col].median()
    df_diabetes[col] = df_diabetes[col].fillna(median_val)

print("\nAfter median imputation, checking for NaN counts:")
display(df_diabetes[columns_with_potential_zeros_as_missing].isnull().sum())

print("\nFirst 5 rows of df_diabetes after handling missing values:")
display(df_diabetes.head())
```

```python
print("Handling outliers in df_diabetes...")

# 1. Identify numerical columns, excluding 'Outcome'
numerical_cols = df_diabetes.columns.drop('Outcome')

# 2-5. Calculate IQR and cap outliers for each numerical column
for col in numerical_cols:
    Q1 = df_diabetes[col].quantile(0.25)
    Q3 = df_diabetes[col].quantile(0.75)
    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    # Cap the outliers
    df_diabetes[col] = np.where(df_diabetes[col] > upper_bound, upper_bound, df_diabetes[col])
    df_diabetes[col] = np.where(df_diabetes[col] < lower_bound, lower_bound, df_diabetes[col])

print("Outliers handled successfully using IQR capping.")
print("First 5 rows of df_diabetes after outlier handling:")
display(df_diabetes.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

print("Applying Min-Max scaling to numerical features of df_diabetes...")

# 1. Create a list of numerical columns to be scaled, excluding 'Outcome'
# The numerical_cols variable from the previous step already contains these columns.
# Re-using numerical_cols from the previous step which was df_diabetes.columns.drop('Outcome')

# 2. Instantiate the MinMaxScaler
scaler = MinMaxScaler()

# Create a temporary DataFrame for scaling to avoid modifying the original during the process, or direct scaling
# Apply Min-Max scaling to the selected numerical columns in df_diabetes using the fit_transform() method.
df_diabetes[numerical_cols] = scaler.fit_transform(df_diabetes[numerical_cols])

print("Min-Max scaling applied successfully.")
print("First 5 rows of df_diabetes after Min-Max scaling:")
display(df_diabetes.head())
```

```python
from sklearn.preprocessing import StandardScaler

print("Applying Standard Scaling to numerical features of df_diabetes...")

# numerical_cols was already defined in the previous step and excludes 'Outcome'
# Re-using numerical_cols from the previous step which was df_diabetes.columns.drop('Outcome')

# Instantiate the StandardScaler
scaler = StandardScaler()

# Apply Standard Scaling to the selected numerical columns in df_diabetes
df_diabetes[numerical_cols] = scaler.fit_transform(df_diabetes[numerical_cols])

print("Standard Scaling applied successfully.")
print("First 5 rows of df_diabetes after Standard Scaling:")
display(df_diabetes.head())
```

## Load Adult Income Dataset

### Subtask:
Download and load the Adult Income dataset into a pandas DataFrame.


```python
import requests
import pandas as pd

# 1. Download the 'adult.csv' dataset
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data'
response = requests.get(url)

if response.status_code == 200:
    with open('adult.csv', 'wb') as f:
        f.write(response.content)
    print("Adult Income dataset downloaded successfully as 'adult.csv'.")
else:
    print(f"Failed to download Adult Income dataset. Status code: {response.status_code}")

# 2. Define the column names
column_names = [
    'age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status',
    'occupation', 'relationship', 'race', 'sex', 'capital_gain', 'capital_loss',
    'hours_per_week', 'native_country', 'income'
]

# 3. Load the 'adult.csv' file into a pandas DataFrame
df_adult = pd.read_csv('adult.csv', names=column_names, header=None)

print("\nAdult Income dataset loaded into df_adult successfully.")

# 4. Display the first 5 rows
print("First 5 rows of df_adult:")
display(df_adult.head())
```

```python
print("1. First 5 rows of df_adult:")
display(df_adult.head())

print("\n2. Information of all columns in df_adult:")
df_adult.info()

print("\n3. Statistical information of numerical columns in df_adult:")
display(df_adult.describe())

print("\n4. Missing values in df_adult:")
display(df_adult.isnull().sum())
```

```python
import numpy as np

print("Handling missing values in df_adult...")

# 1. Identify all columns in df_adult that have an object data type
object_cols = df_adult.select_dtypes(include='object').columns

# 2. For each identified object column, check if the string ' ?' exists as a value.
# If it does, replace all occurrences of ' ?' with np.nan.
for col in object_cols:
    if ' ?' in df_adult[col].unique():
        df_adult[col] = df_adult[col].replace(' ?', np.nan)
        print(f"Replaced ' ?' with NaN in column: {col}")

# 3. After replacing ' ?' with np.nan, display the count of missing values (NaNs)
# for all columns using df_adult.isnull().sum() to confirm the replacement.
print("\nMissing values after replacing ' ?' with NaN:")
display(df_adult.isnull().sum())

# 4. For each column where NaNs were introduced, impute these NaN values using the mode.
for col in object_cols:
    if df_adult[col].isnull().sum() > 0:
        mode_val = df_adult[col].mode()[0] # .mode() returns a Series, take the first element
        df_adult[col] = df_adult[col].fillna(mode_val)
        print(f"Imputed NaN values with mode ({mode_val}) in column: {col}")

# 5. Finally, display the count of missing values again to ensure all NaNs are handled,
# and display the first 5 rows of the updated df_adult DataFrame.
print("\nMissing values after mode imputation:")
display(df_adult.isnull().sum())

print("\nFirst 5 rows of df_adult after handling missing values:")
display(df_adult.head())
```

```python
print("Cleaning and encoding categorical features in df_adult...")

# 1. Identify all columns in df_adult that have an object data type.
object_cols = df_adult.select_dtypes(include='object').columns
print(f"Original object columns: {list(object_cols)}")

# 2. For each identified categorical column, remove any leading or trailing whitespace.
for col in object_cols:
    df_adult[col] = df_adult[col].str.strip()
    print(f"Trimmed whitespace in column: {col}")

# 3. Encode the 'income' column
print("\nEncoding 'income' column...")
income_mapping = {'<=50K': 0, '>50K': 1}
df_adult['income'] = df_adult['income'].map(income_mapping)
print("Value counts of encoded 'income' column:")
display(df_adult['income'].value_counts())

# Remove 'income' from object_cols as it's now numerical
object_cols = object_cols.drop('income', errors='ignore')

# 4. Apply one-hot encoding to the remaining categorical columns
print("\nApplying one-hot encoding to remaining categorical columns...")
df_adult = pd.get_dummies(df_adult, columns=object_cols, drop_first=True)

print("Categorical features encoded successfully.")

# 5. Display the first 5 rows of the updated df_adult DataFrame
print("\nFirst 5 rows of df_adult after categorical encoding:")
display(df_adult.head())
```

```python
import numpy as np

print("Handling outliers in df_adult...")

# 1. Identify numerical columns, excluding 'income'
# The 'income' column was already transformed into numerical (0 or 1) and is the target variable.
# Other columns are either numerical (age, fnlwgt, education_num, capital_gain, capital_loss, hours_per_week)
# or one-hot encoded binary columns, which typically aren't treated for outliers in this manner.

numerical_cols_for_outliers = [
    'age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week'
]

# 2-6. Calculate IQR and cap outliers for each identified numerical column
for col in numerical_cols_for_outliers:
    Q1 = df_adult[col].quantile(0.25)
    Q3 = df_adult[col].quantile(0.75)
    IQR = Q3 - Q1

    upper_bound = Q3 + 1.5 * IQR
    lower_bound = Q1 - 1.5 * IQR

    # Cap the outliers
    df_adult[col] = np.where(df_adult[col] > upper_bound, upper_bound, df_adult[col])
    df_adult[col] = np.where(df_adult[col] < lower_bound, lower_bound, df_adult[col])
    print(f"Outliers capped for column: {col}")

print("Outliers handled successfully using IQR capping for specified numerical columns.")

# 7. Display the first 5 rows of df_adult after outlier handling
print("\nFirst 5 rows of df_adult after outlier handling:")
display(df_adult.head())
```

```python
from sklearn.preprocessing import MinMaxScaler

print("Applying Min-Max scaling to numerical features of df_adult...")

# 1. Define a list of numerical columns to be scaled.
# These are the same columns identified for outlier handling.
numerical_cols_for_scaling = [
    'age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week'
]

# 2. Instantiate a MinMaxScaler object.
scaler = MinMaxScaler()

# 3. Apply Min-Max scaling to the selected numerical columns in df_adult
df_adult[numerical_cols_for_scaling] = scaler.fit_transform(df_adult[numerical_cols_for_scaling])

print("Min-Max scaling applied successfully.")

# 4. Display the first 5 rows of the updated df_adult DataFrame
print("First 5 rows of df_adult after Min-Max scaling:")
display(df_adult.head())
```

```python
from sklearn.preprocessing import StandardScaler

print("Applying Standard Scaling to numerical features of df_adult...")

# numerical_cols_for_scaling was already defined in the previous step
# It contains: ['age', 'fnlwgt', 'education_num', 'capital_gain', 'capital_loss', 'hours_per_week']

# Instantiate the StandardScaler
scaler = StandardScaler()

# Apply Standard Scaling to the selected numerical columns in df_adult
df_adult[numerical_cols_for_scaling] = scaler.fit_transform(df_adult[numerical_cols_for_scaling])

print("Standard Scaling applied successfully.")
print("First 5 rows of df_adult after Standard Scaling:")
display(df_adult.head())
```
